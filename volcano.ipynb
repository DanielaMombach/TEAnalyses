## Volcano plot

source: https://www.reneshbedre.com/blog/volcano.html

```python
import pandas as pd
```

```python
from bioinfokit import analys, visuz
```

```python
df = pd.read_csv('srr_cispla/TEtranscripts_results/A2780_deseq_all.csv')
```

```python
df
```

```python
df['log2FoldChange'].dtype
```

### A2780


```python
visuz.GeneExpression.volcano(df=df, lfc='log2FoldChange', pv='padj', color=("#00239CFF", "grey", "#E10600FF"), lfc_thr=(1, 1), pv_thr=(0.05, 0.05), valpha=0.5, dotsize=25,  xlm=(-3,2,0.5), ylm=(0,101,20), sign_line=True, show=False)
```

### Acis

```python
visuz.GeneExpression.volcano(df=df, lfc='log2FoldChange', pv='padj', color=("#00239CFF", "grey", "#E10600FF"), lfc_thr=(1, 1), pv_thr=(0.05, 0.05), valpha=0.5, dotsize=25, sign_line=True, markerdot='*', xlm=(-3,2,0.5), ylm=(0,101,20), show=False)
```
