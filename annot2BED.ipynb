##### but first in shell: $ sed 's/\t/,/g' gencode.v42.annotation.gtf | tail -n +6 > gencode.v42.annotation.csv

```python
import pandas as pd
```

```python
gene = pd.read_csv('gencode.v42.annotation.csv', header=None)
```

```python
gene
```

```python
gene=gene.loc[gene[2].str.startswith('gene')]
```

```python
gene[2].unique() #check
```

```python
gene[0].value_counts() #count how many of each type
```

```python
gene[0].replace('^chr', 'cro', regex=True) #replace a pattern
```

```python
gene[8].str.split('"',expand=True)
```

```python
gene[9] = gene[8].str.split('"', expand=True)[1]
gene
```

```python
BED = gene[[0, 3, 4, 9, 5, 6]]
gene
```

### REdiscoverTE annotation

```python
df = pd.read_csv('REdiscoverTE/original/REdiscoverTE/rollup_annotation/rmsk_annotation.csv')
```

```python
df['idx'].str.split(',') #split
```

```python
df['nova'] = df['idx'].str.split(',') #save split
df
```

```python
df.explode('nova', ',') #explode
```

```python
df = df.explode('nova', ',') #save explode
```

```python
df['chr'] = df['nova'].str.split('__', expand = True)[0]
df['coo'] = df['nova'].str.split('__', expand = True)[1]
## or df[ ['chr', 'coo'] ] = df['nova'].str.split('__', expand = True)
```

```python
df['coo'].str.split('_', expand=True)
```

```python
df['start'] = df['coo'].str.split('_', expand = True)[0]
df['end'] = df['coo'].str.split('_', expand = True)[1]
```

```python
BED = df[ ['chr', 'start', 'end', 'repName', 'repClass'] ]
```

```python
BED['chr'] = BED['chr'].apply(lambda x: f'chr{x}') #insert chr before "x"
```

```python
BED
```

```python
BED.to_csv('rmsk_annotation.bed', sep='\t', index=False, header=False)
```

```python
rmsk=pd.read_table('rmsk_annotation.bed', header=None)
```

```python
rmsk_plus
```
